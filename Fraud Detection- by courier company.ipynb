```python
new=pd.merge(d,b,on=["Warehouse Pincode","Customer Pincode"],how="inner")
#new[new["Zone_x"]!=new["Zone_y"]]
new_1=pd.merge(new,y,on="Order ID",how="inner")
def x(pi):
    if (pi["total_weight"]>0.0)and (pi["total_weight"]<=0.5):
        group=0.5
        return group
      
    elif (pi["total_weight"]>0.5)and (pi["total_weight"]<=1.0):
        group=1.0
        return group
    
    elif (pi["total_weight"]>1.0)and (pi["total_weight"]<=1.5):
        group=1.5
        return group
    elif (pi["total_weight"]>1.5)and (pi["total_weight"]<=2.0):
        group=2
        return group
    elif (pi["total_weight"]>2.0)and (pi["total_weight"]<2.5):
        
        group=2.5
        return group
    elif (pi["total_weight"]>2.5)and (pi["total_weight"]<3.0):
        group=3
        return group
    elif (pi["total_weight"]>3.0)and (pi["total_weight"]<=3.5):
        group=3.5
        return group
    elif pi["total_weight"]==0.5:
        group=1.0
        return group
    else:
        group=4.0
        return group
    
new_1["final"]=new_1.apply(x,axis=1)
new_1
#new_1["Zone_y"].value_counts()
#new_1.isnull().sum()
new_1.dtypes
#new_1["final"]=new_1["final"].astype("int64")
new_1["x"]=new_1["final"]/0.5
```

```python
def rates(r):
    if(r["Zone_y"]=="d")and (r["Type of Shipment"]=="Forward charges"):
        if(r["final"]==0.5):
            group=(r["x"]*45.4)
            return group
    
        elif(r["final"]>0.5):
            group=(45.4+((r["x"]-1.0)*44.8))
            return group
    elif(r["Zone_y"]=="d")and (r["Type of Shipment"]=="Forward and RTO charges"):
        if(r["final"]==0.5):
            group=(r["x"]*(45.4+41.3))
            return group
    
        elif(r["final"]>0.5):
            group=((45.4+41.3)+((r["x"]-1.0)*(44.8+44.8)))
            return group
        
        
    elif(r["Zone_y"]=="b")and (r["Type of Shipment"]=="Forward charges"):
        if(r["final"]==0.5):
            group=(r["x"]*33)
            return group
    
        elif(r["final"]>0.5):
            group=(33+((r["x"]-1.0)*28.3))
            return group
    elif(r["Zone_y"]=="b")and (r["Type of Shipment"]=="Forward and RTO charges"):
        if(r["final"]==0.5):
            group=(r["x"]*(33+20.5))
            return group
    
        elif(r["final"]>0.5):
            group=((33+20.5)+((r["x"]-1.0)*(28.3+28.3)))
            return group
        
    elif(r["Zone_y"]=="e")and (r["Type of Shipment"]=="Forward charges"):
        if(r["final"]==0.5):
            group=(r["x"]*56.6)
            return group
    
        elif(r["final"]>0.5):
            group=(56.6+((r["x"]-1.0)*55.5))
            return group
    elif(r["Zone_y"]=="e")and (r["Type of Shipment"]=="Forward and RTO charges"):
        if(r["final"]==0.5):
            group=(r["x"]* (56.6+50.7))
            return group
    
        elif(r["final"]>0.5):
            group=( (56.6+50.7)+((r["x"]-1.0)*(55.5+55.5)))
            return group
    #(56.6+50.7)+(55.5+55.5)
                   
new_1["Company_x_amount"]=new_1.apply(rates,axis=1)
new_1
```

```python
def c_1(w):
    if (w["Billing Amount (Rs.)"])>(w["Company_x_amount"]):
        group="Extra Charged"
        return group
    elif(w["Billing Amount (Rs.)"])<(w["Company_x_amount"]):
        
        group="Less charged"
        return group
    else:
        group="correctly charged"
        return group
new_1["Company_x_diff"]=new_1.apply(c_1,axis=1)
```
